(1) 常見評估指標
1. Accuracy(準確率):
  * 定義：衡量模型整體預測正確的比例
  * 公式：(TP + TN) / (TP + TN + FP + FN)
2. Precision(精確率):
  * 定義：在所有被模型預測為正類的案例中，實際為正類的比例
  * 公式：TP / (TP + FP)
3. Recall(召回率):
  * 定義：在所有實際為正類的案例中，模型成功預測出來的比例
  * 公式：TP / (TP + FN)
4. F1-score:
  * 定義：Precision與召回率間的平衡指標，當類別不平衡時特別重要
  * 公式：(2 * Precision * Recall) / (Precision + Recall)
  * 公式可再展開
5. ROC-AUC
  * 衡量模型區分正負類的能力
  * ROC曲線下的面積，用來衡量二分類模型的性能。AUC越接近1，模型越好

 **Please implement a function to calculate the accuracy, precision, recall, and F1-score for each level. Based on the results, provide your interpretation and opinion on the outcomes.**

```
import numpy as np
import pandas as pd


def evaluation_metrics(confusion_matrix):
  metrics = {}
  levels = confusion_matrix.index

  for level in levels:

    # 預測為lavel，且實際也為level
    tp = confusion_matrix.loc[level, level]
    # 預測為lavel，且實際不為level
    fp = confusion_matrix[level].sum() - tp
    # 預測不為lavel，且實際為level
    fn = confusion_matrix.loc[level].sum() - tp
    # 預測不為lavel，且實際也不為level
    tn = confusion_matrix.values.sum() - (tp + fp + fn)

    accuracy = (tp + tn) / confusion_matrix.values.sum()
    precision = tp / (tp + fp) if (tp + fp) != 0 else 0
    recall = tp / (tp + fn) if (tp + fn) != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

    metrics[level] = {
      "Accuracy": accuracy,
      "Precision": precision,
      "Recall": recall,
      "F1-score": f1_score,
    }

  return(metrics)

metrics = evaluation_metrics(df)
metrics_df = pd.DataFrame(metrics)
print(metrics_df)
```


## 2-1. Define a *funtion* aims to know the weekly (7 days) Stock **Close value** trend. (80%)
- Implement a *funtion* that determines the weekly (7 days) Stock **Close value** trend using the Stock data and **then record the trend in the News data in a new column "label" with 0, 1 and -1**. For example,
  - On 2016-06-13, the market closed at 17732.480469. Seven days later, 2016-06-20, the market closed **higher** at 17804.869141. In this scenario, all entries corresponding to 2016-06-13 in the News data will be **marked 1** in the "label" column. Dates 2016-06-14 and 2016-06-15 will also be marked 1 because 2016-06-21 and 2016-06-22 closed higher, respectively.
  - On the other hand, 2016-06-17 will be **marked 0** because 2016-06-24 (7 days later) closed **lower**.
  - If a given date does not have a corresponding date for 7 days later, the given date will be **marked -1**.




```
import datetime

def label(stock_df):

  stock_df['Date'] = pd.to_datetime(stock_df['Date'])
  # key為'Date';value為'Close'
  stock_dict = stock_df.set_index('Date')['Close'].to_dict()

  labels_dict = {}

  for stock_date in stock_dict.keys():
    seven_days_date = stock_date + datetime.timedelta(days=7)

    if seven_days_date in stock_dict:
      # 7天後的收盤價比當天高
      if stock_dict[seven_days_date] > stock_dict[stock_date]:
        labels_dict[stock_date] = 1
      # 7天後的收盤價比當天低
      elif stock_dict[seven_days_date] < stock_dict[stock_date]:
        labels_dict[stock_date] = 0
      # 7天後的收盤價和當天一樣
      else:
        labels_dict[stock_date] = -1
    # 7天後的日期不存在
    else:
      labels_dict[stock_date] = -1

  return labels_dict
```

```
labels_dict = label(stock_df)
stock_df['Label'] = stock_df['Date'].map(labels_dict)
stock_df.head(15)
```

2. 標籤對齊：

  * [CLS] 和 [SEP] 這些特殊的標記不對應任何詞語被標註為 -100

> 在分詞後，將每個詞語的標籤正確地對應到它的每個子詞
> 且對於特殊的子詞或標記，給予標籤 -100，讓模型在訓練時忽略它們



---

* 範例：

  * 原始句子：["EU", "rejects", "German", "lamb"]

## 3-6:  Define Evaluation Metrics




* **TODO3 & TODO4:  Complete the compute_metrics Function**

*   In Named Entity Recognition (NER) tasks, the model outputs predicted labels for each token, and the true labels must be aligned with the predictions. After alignment, the labels and predictions need to be flattened for further evaluation metrics calculation, such as precision, recall, and F1-score.

*  Task Objective:
Complete the missing parts in the compute_metrics function to achieve the following:
>


1.   Align true labels and predictions:
*   Ignore padding tokens (-100).
*   Map numerical labels to their corresponding names using the provided id_to_label.
2.   Flatten nested lists:
*   Convert the aligned labels and predictions into one-dimensional lists for evaluation.


```
def compute_metrics(pred):

  import numpy as np
  from sklearn.metrics import classification_report

  # Extract predictions and true labels
  predictions, labels = pred.predictions, pred.label_ids

  # Convert logits to predicted label indices
  predictions = np.argmax(predictions, axis=2)

  # Align true labels and predictions, ignoring padding (-100)
  true_labels = []
  true_predictions = []

  for i in range(len(labels)):
    # 取得真實標籤
    label = labels[i]
    # 取得預測標籤
    prediction = predictions[i]

    for true_label, pred_label in zip(label, prediction):

      if true_label != -100:
        # 使用 id_to_label 將數字標籤轉換為標籤名稱
        true_labels.append(id_to_label[true_label])
        true_predictions.append(id_to_label[pred_label])

  # 扁平化列表，準備計算評估指標
  flat_labels = true_labels
  flat_predictions = true_predictions

  # Generate classification report
  report = classification_report(flat_labels, flat_predictions, output_dict=True)

  return report
```

```
from sklearn.metrics import accuracy_score
# Load model and tokenizer
model = AutoModelForTokenClassification.from_pretrained("./ner_model")
tokenizer = AutoTokenizer.from_pretrained("./ner_model")

# Load test dataset
dataset = load_dataset("wikiann", "en")

# Extract a subset of the test dataset
test_dataset = dataset["test"][:1000]

# Extract tokens and labels from the test data
tokens_list = test_dataset["tokens"]
labels_list = test_dataset["ner_tags"]
label_mapping = dataset["train"].features["ner_tags"].feature.names

# Initialize lists to store true and predicted labels
true_labels = []
pred_labels = []

# Import time to estimate duration
import time

# Estimate the time per sample and total time
def estimate_time_per_sample(model, tokenizer, tokens):
    start_time = time.time()
    inputs = tokenizer(tokens, is_split_into_words=True, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        _ = model(**inputs)
    return time.time() - start_time

# Sample a few tokens to estimate time per sample
sample_tokens = tokens_list[:5]
average_time_per_sample = sum(estimate_time_per_sample(model, tokenizer, tokens) for tokens in sample_tokens) / len(sample_tokens)
total_estimated_time = average_time_per_sample * len(tokens_list)

# Display estimated time to user
print(f"Estimated time to process {len(tokens_list)} samples: {total_estimated_time:.2f} seconds")

# Model predictions and alignment of true and predicted labels
start_time = time.time()
for tokens, ner_tags in zip(tokens_list, labels_list):
    # Encode tokens for model input
    inputs = tokenizer(tokens, is_split_into_words=True, return_tensors="pt", truncation=True, padding=True)

    # Get model outputs
    with torch.no_grad():
        outputs = model(**inputs)

    # Obtain predicted labels
    logits = outputs.logits
    predictions = torch.argmax(logits, dim=2).squeeze().tolist()

    # Get word IDs to align predictions with the original tokens
    word_ids = inputs.word_ids()

    # Align predictions with original tokens and filter out special tokens
    previous_word_idx = None
    for word_idx, prediction in zip(word_ids, predictions):
        if word_idx is not None and word_idx != previous_word_idx:  # Ignore special tokens and subwords
            if 0 <= prediction < len(label_mapping):  # Ensure prediction is valid
                pred_labels.append(label_mapping[prediction])
                true_labels.append(label_mapping[ner_tags[word_idx]])  # Align with the original token's label
            previous_word_idx = word_idx

# Generate detailed classification report
classification_report_str = classification_report(true_labels, pred_labels, target_names=label_mapping, zero_division=0)

print("Classification Report:\n")
print(classification_report_str)
# Calculate accuracy
accuracy = accuracy_score(true_labels, pred_labels)

print(f"Accuracy: {accuracy}")

# Display total execution time
end_time = time.time()
print(f"Total execution time: {end_time - start_time:.2f} seconds")
```
